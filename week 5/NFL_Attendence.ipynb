#  Comparison between the Teams Chiefs and Giants

```
Chiefs  = attendance[attendance['team_name']=='Chiefs']
Giants = attendance[attendance['team_name']=='Giants']
```

```
new = pd.concat([Chiefs, Giants])
```

# Comparison of both teams attendance in year 2000 and 2019

```
new_2000 = new[new['year']==2000]
new_2019 = new[new['year']==2019]

plt.figure(figsize= (20,7))
sns.barplot(x = 'week',y = 'weekly_attendance',data = new_2000, hue = 'team_name', palette='husl')

plt.figure(figsize= (20,7))
sns.barplot(x = 'week',y = 'weekly_attendance',data = new_2019, hue = 'team_name', palette='husl')
```

```
plt.style.use('seaborn-dark-palette')
plt.figure(figsize= (20,7))
Chiefs.groupby('year',sort = True).weekly_attendance.agg('max').plot(kind = 'bar')
plt.title(('weekly_maximum_attendance of the Chiefs in each year'),fontsize = 15)
```

```
plt.style.use('seaborn-dark-palette')
plt.figure(figsize= (20,7))
Chiefs.groupby('year',sort = True).weekly_attendance.agg('min').plot(kind = 'bar',color = '#524939')
plt.title(('weekly_minimum_attendance of the Chiefs in each year'),fontsize = 15)
```
